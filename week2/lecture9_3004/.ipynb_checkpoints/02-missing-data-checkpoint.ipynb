```python
# importar as lib que vou precisar
import random
import numpy as np
import pandas as pd
import seaborn as sns

# reproducibility - essa semente vai deixar os meus valores fixos em qualquer compilação desse código
#A função de semente é usada para salvar o estado de uma função aleatória, para que possa gerar os mesmos números aleatórios em várias execuções do código na mesma máquina ou em máquinas diferentes (para um valor de semente específico). 
random.seed(0)
```

## The Palmer penguins dataset

```python
# carregando a base dos pinguins e armazenando na variável df
df = sns.load_dataset('penguins')
```

```python
# mostrando os 5 primeiros casos/id
df.head()
```

### Features
- `species`: a factor denoting penguin species (Adélie, Chinstrap and Gentoo)
- `island`: a factor denoting island in Palmer Archipelago, Antarctica (Biscoe, Dream or Torgersen)
- `bill_length_mm`: a number denoting bill length (millimeters)
- `bill_depth_mm`: a number denoting bill depth (millimeters)
- `flipper_length_mm`: an integer denoting flipper length (millimeters)
- `body_mass_g`: an integer denoting body mass (grams)
- `sex`: a factor denoting penguin sex (female, male)

### Encoding

```python

# use pd.concat to join the new columns with your original dataframe
df = pd.concat([df,pd.get_dummies(df[['island','species']], prefix=['island', 'species'])], axis=1)
df = pd.concat([df,pd.get_dummies(df['sex'], prefix='sex')],axis=1)

# now drop the original 'country' column (you don't need it anymore)
df.drop(['island', 'species', 'sex'],axis=1, inplace=True)
```

```python
df
```

### Missing data
#### Numerical

```python
# simulating missing values
column = 'body_mass_g'
# vou cortar 40% dos meus dados
missing_pct = .4

missing_rows = np.random.choice(
    range(df[column].shape[0]),
    size=int(df[column].shape[0] * missing_pct),
    replace=False
)

df.loc[missing_rows, column] = np.NaN
```

```python
df
```

#### Eliminate rows with missing data

```python
# remove os valores nulos (valores ausentes) do DataFrame , deixando cair as linhas ou colunas que contêm os valores nulos
df.dropna()
```

#### Eliminate columns with missing data

```python
# só colocar o axis=1
df.dropna(axis=1)
```

#### Imputation: mean, median, mode

```python
# fazer imputação de dados usando média, mediana ou moda
# df.illna() substitui os valores NaN em DataFrame por algum determinado valor.
# aqui ela vai substituir pela média
#df.fillna(df.mean())
# aqui ela vai substituir pela mediana
# df.fillna(df.median())
# aqui ela vai substituir pela moda
df.fillna(df.mode().squeeze())
```
