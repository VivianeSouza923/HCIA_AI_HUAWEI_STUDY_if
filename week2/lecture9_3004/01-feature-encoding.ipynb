## The Palmer penguins dataset

```python
# usando a lib seaborn, eu vou pegar o dataset dos pinguins e armazenar em df
df = sns.load_dataset('penguins')
```

### Features
- `species`: a factor denoting penguin species (Adélie, Chinstrap and Gentoo)
- `island`: a factor denoting island in Palmer Archipelago, Antarctica (Biscoe, Dream or Torgersen)
- `bill_length_mm`: a number denoting bill length (millimeters)
- `bill_depth_mm`: a number denoting bill depth (millimeters)
- `flipper_length_mm`: an integer denoting flipper length (millimeters)
- `body_mass_g`: an integer denoting body mass (grams)
- `sex`: a factor denoting penguin sex (female, male)

### Ecoding

```python
# o pd.get_dumnies() irá servir para converter variáveis categóricas em variáveis dummy / indicadoras.
# pd.get_dumies(data, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)
# no nosso caso, estamos usando só o data e o prefix. O dado que a gente tá pegando são as colunas de island e species do dataframe df
# e o prefixo a gente tá chjamdno, na ordem, island e species

#one-hot encondig
pd.get_dummies(df[['island','species']], prefix=['island', 'species']) # drop_first=True to dummy encoding


# já para dummy enconding:
# pd.get_dummies(df[['island','species']], prefix=['island', 'species'], drop_first=True)
# usando o drop_first=True, se deve-se obter k-1 dummies de k níveis categóricos removendo o primeiro nível.
```

#### 

```python
# use pd.concat to join the new columns with your original dataframe
# aqui a gente vai adicionar os dados codificados em one-hot encoding no dataframe (df), adicionando as colunas (axis=1)
df = pd.concat([df,pd.get_dummies(df[['island','species']], prefix=['island', 'species'])], axis=1)
# em seguida, eu quero adicionar mais uma nova condificação que eu fiz (pegar o sexo e transformar em vetor e rotular, prefixar como "sex" mesmo)
df = pd.concat([df,pd.get_dummies(df['sex'], prefix='sex')],axis=1)

# now drop the original 'country' column (you don't need it anymore)
df.drop(['island', 'species', 'sex'],axis=1, inplace=True)
```

### Features
Both hour.csv and day.csv have the following fields, except hr which is not available in day.csv

- instant: record index
- dteday : date
- season : season (1:springer, 2:summer, 3:fall, 4:winter)
- yr : year (0: 2011, 1:2012)
- mnth : month ( 1 to 12)
- hr : hour (0 to 23)
- holiday : weather day is holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
- weekday : day of the week
- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
+ weathersit : 
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp : Normalized temperature in Celsius. The values are divided to 41 (max)
- atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)
- hum: Normalized humidity. The values are divided to 100 (max)
- windspeed: Normalized wind speed. The values are divided to 67 (max)
- casual: count of casual users
- registered: count of registered users
- cnt: count of total rental bikes including both casual and registered
