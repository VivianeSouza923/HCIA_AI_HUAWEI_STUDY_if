##Series

```
# importar as libs
import numpy as np
import pandas as pd
```

```
# criação de array aleatória com 3 números dentro de 5 a 49
arr = np.random.randint(5, 50, 3)
# mostra array
arr
```

```
# criar serie
# serieCriar = pd.Series(arrayCriada)
series = pd.Series(arr)
# mostra
series
```

```
# passa-se os rótulos da serie 
labels = ['a', 'b', 'c']
# passa o label na criação de series mesmo, só acrescentar o parâmetro
series = pd.Series(arr, labels)
#mostra
series
```

```
#localiza de forma simples através de label
series['a']
```

```
states = ['CE', 'RN', 'PI', 'MA']
sales_per_state1 = np.random.randint(100, 500, 4)
sales_per_state1
```

```
sales_per_state2 = np.random.randint(100, 500, 4)
sales_per_state2
```

```
# criando a serie com labels já criados anteriormente
series_sales1 = pd.Series(sales_per_state1, states)
series_sales1
```

```
series_sales2 = pd.Series(sales_per_state2, states)
series_sales2
```

```
series_sales1['MA']
```

```
series_sales2[1]
```

```
# soma de serires
series_sales1 + series_sales2
```

```
# criação de dicionário
dic = {'CE': 100, 'RN': 150, 'MA': 120}
dic
```

```
# adiciona o dicionário nas series
series = pd.Series(dic)
series
```
