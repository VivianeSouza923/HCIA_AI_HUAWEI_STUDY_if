## Matplotlib
* Plotting library for Python

```
# Libs que vou precisar

import numpy as np
import pandas as pd
# lib mais usada para mostrar gr치ficos em Python
import matplotlib.pyplot as plt
```

```
# criar uma array usando arange, colocando de 1 a 50
x = np.arange(1, 50)
# colocar na vari치vel x para dar uma olhada
x
```

```
# criar o y agora (escolhi fazer o log de x)
y = np.log(x)
# dar olhada em y
y
```

```
# j치 d치 pra mostrar
plt.plot(x, y)

plt.show()
```

```
# arange cria uma array do -20 a 20 (lembrando que este 칠 exclusivo)
x2 = np.arange(-20, 20)
x2
```

```
# pro y2 eu vou criar uma fun칞칚ozinha (em fun칞칚o de x2 mesmo)
y2 = 0.5 * x2 ** 2 + 0.5* x2 + 0.5
y2
```

```
# plotar
plt.plot(x2, y2)
```

```
# plotar os dados com a cor da linha que eu quero
# plt.plot(x2,y2,'letra_da_inicial_cor_que_quero')
plt.plot(x2, y2, 'y')
```

```
# quero a linha pontilhada? d치 pra fazer. s칩 colocar a letra da cor aompanhada de --, por exemplo r-- --> red pontilhado.
plt.plot(x, y, 'r--')
```

```
# quero decidir a dimens칚o da figura que vai sair o meu plot
# usa s칩 o .figure(figsize=(largura, altura))
plt.figure(figsize=(10, 6))
# misturando os dois gr치ficos
plt.plot(x, y, 'r', x2, y2, 'b--')
```

```
plt.plot(x, y, 'r', x2, y2, 'b--')
# se eu quiser colocar legenda, s칩 usar o .legend([]). se tiver mais de uma, s칩 separar por v칤rgulas.
plt.legend(['log', 'x**2'])
```

```
plt.plot(x, y)
#r칩tulo
plt.xlabel('x')
#r칩tulo
plt.ylabel('log(x)')
# botar o t칤tulo, posso escolher o tamanho da fonte para adequar ao tamanho da minha figura.
plt.title('Plot', fontsize=20)

plt.show()
```

###Subplots

```
# criar figura
# as dimens칫es da figura podem ser passadas
#plt.figure(figsize=(base, altura))
plt.figure(figsize=(15, 5))

# aqui a gente vai SUBplotar algo, ou seja, temos a figura e desejamos subplotar, plotar algo, o dado dentro dessa figura, entende?
#plt.subplot(numero_de_linhas, numero_de_colunas, index)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
#limite do x, at칠 onde ele vai. como ele vai para a direita, s칩 precisa entrar como par칙metro (right=valor_que_eu_quero)
plt.xlim(right=60)
#limites de y (lembrando que o menor valor 칠 exclusivo)
plt.ylim(bottom=-5, top=8)
#botar os valores x, usei essa gera칞칚o de valores atrav칠s de arange
plt.xticks(np.arange(0, 61, 5))

# mesma coisa do anterior, a diferen칞a, 칩bvio 칠 a segunda figura vai estar no index 2
plt.subplot(1, 2, 2)
plt.plot(x2, y2, 'b--')
# como ele t치 no index 2, ou seja, tem gente mais pra frente, 칠 importante botar aqui n칚o s칩 um, mas os dois, o "left" e o "right"
plt.xlim(left=-30, right=60)

# s칩 mostrar agoraaa
plt.show()
```

## Seaborn
* Statistics plots - gr치ficos estat칤sticos
* Built on top of Matplotlib

```
# mostra uma no칞칚ozinha de base
data_heart.head()
```

```
# importa a lib, ou seja, seaborn, que 췂e quem a gente quer usar
import seaborn as sns
```

```
#displot - veja como 칠 rodando a c칠lula (POR PADR츾O, sem informar o kind, ele sai assim com esses ret칙ngulos bonitinhos). como ficam os dados de idade?
# o hist, histograma 칠 o padr칚o, esses ret칙ngulos.
sns.displot(data_heart['age'])
```

```
# informando o kind, a gente pode variar a representa칞칚o desses dados de muitos jeitos
sns.displot(data_heart['age'], kind='kde')
#sns.displot(data_heart['age'], kind='ecdf')
```

```
# quero a  minha representa칞칚o padr칚o, mas quero um kde por cima? s칩 adicionar kde=True
#sns.displot(data_heart['age'], kde=True)

# quero a mesma coisa, mas com o rug por cima?
sns.displot(data_heart['age'], rug=True)
```

```
# o bins passa quantos intervalos ir칚o ter a representa칞칚o. Falando de uma forma beem bruta, ele vai dar 15 ret칙ngulos (dividir os dados nisso), nem mais, nem menos.
# s칩 d치 pra fazer essa quantiza칞칚o com kind='hist', ou seja, os ret칙ngulos, o padr칚o que n칚o precisa colocar nada.
sns.displot(data_heart['age'], bins=15)
```

```
# posso trocar a cor desses ret칙ngulos, escolher a altura do plot, que ser치 de 8 unidades.
sns.displot(data_heart['age'], bins=20, color='r', height=8)
```

```
# mostra
data_heart.head()
```

```
# agora a gente vai usar o countplot, plotar de acordo com quantidade(?)
# exatamente, contagem de valores categ칩ricos
# sns.countplot(data=nome_da_base, x='nome_da_coluna')
sns.countplot(data=data_heart, x='sex')
#perceba que ele est치 contando o n칰mero de zeros, o n칰mero de uns.
```

```
# agora com uma coluna com dados n칚o bin치rios
sns.countplot(data=data_heart, x='cp')
#criar uma label pra fazer uma gra칞a
plt.xlabel('Chest Pain Type')
```

```
# o hue n칚o adiciona somente uma legenda. voc칡 pode adicionar a coluna  para o "hue" que deseja
# o que est치 acontecendo aqui? o hue(sexo) est치 mostrando sua quantiza칞칚o para cada cp(0,1,2 e 3)
sns.countplot(data=data_heart, x='cp', hue='sex')
```

```
# mostra
data_heart.head()
```

```
# boxplot, sa칤mos de countplot. S칩 passar a coluna como uma vari치vel y, pra ela ficar em p칠zinha. Se passar como x, ela fica deitada, saca?
sns.boxplot(data=data_heart, y='age')
```

```
# boxplot de idade por sexo
sns.boxplot(data=data_heart, y='age', x='sex')
```
