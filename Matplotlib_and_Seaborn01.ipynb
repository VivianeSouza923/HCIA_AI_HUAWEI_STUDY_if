## Matplotlib
* Plotting library for Python

```
# Libs que vou precisar

import numpy as np
import pandas as pd
# lib mais usada para mostrar gráficos em Python
import matplotlib.pyplot as plt
```

```
# criar uma array usando arange, colocando de 1 a 50
x = np.arange(1, 50)
# colocar na variável x para dar uma olhada
x
```

```
# criar o y agora (escolhi fazer o log de x)
y = np.log(x)
# dar olhada em y
y
```

```
# já dá pra mostrar
plt.plot(x, y)

plt.show()
```

```
# arange cria uma array do -20 a 20 (lembrando que este é exclusivo)
x2 = np.arange(-20, 20)
x2
```

```
# pro y2 eu vou criar uma funçãozinha (em função de x2 mesmo)
y2 = 0.5 * x2 ** 2 + 0.5* x2 + 0.5
y2
```

```
# plotar
plt.plot(x2, y2)
```

```
# plotar os dados com a cor da linha que eu quero
# plt.plot(x2,y2,'letra_da_inicial_cor_que_quero')
plt.plot(x2, y2, 'y')
```

```
# quero a linha pontilhada? dá pra fazer. só colocar a letra da cor aompanhada de --, por exemplo r-- --> red pontilhado.
plt.plot(x, y, 'r--')
```

```
# quero decidir a dimensão da figura que vai sair o meu plot
# usa só o .figure(figsize=(largura, altura))
plt.figure(figsize=(10, 6))
# misturando os dois gráficos
plt.plot(x, y, 'r', x2, y2, 'b--')
```

```
plt.plot(x, y, 'r', x2, y2, 'b--')
# se eu quiser colocar legenda, só usar o .legend([]). se tiver mais de uma, só separar por vírgulas.
plt.legend(['log', 'x**2'])
```

```
plt.plot(x, y)
#rótulo
plt.xlabel('x')
#rótulo
plt.ylabel('log(x)')
# botar o título, posso escolher o tamanho da fonte para adequar ao tamanho da minha figura.
plt.title('Plot', fontsize=20)

plt.show()
```

###Subplots

```
# criar figura
# as dimensões da figura podem ser passadas
#plt.figure(figsize=(base, altura))
plt.figure(figsize=(15, 5))

# aqui a gente vai SUBplotar algo, ou seja, temos a figura e desejamos subplotar, plotar algo, o dado dentro dessa figura, entende?
#plt.subplot(numero_de_linhas, numero_de_colunas, index)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
#limite do x, até onde ele vai. como ele vai para a direita, só precisa entrar como parâmetro (right=valor_que_eu_quero)
plt.xlim(right=60)
#limites de y (lembrando que o menor valor é exclusivo)
plt.ylim(bottom=-5, top=8)
#botar os valores x, usei essa geração de valores através de arange
plt.xticks(np.arange(0, 61, 5))

# mesma coisa do anterior, a diferença, óbvio é a segunda figura vai estar no index 2
plt.subplot(1, 2, 2)
plt.plot(x2, y2, 'b--')
# como ele tá no index 2, ou seja, tem gente mais pra frente, é importante botar aqui não só um, mas os dois, o "left" e o "right"
plt.xlim(left=-30, right=60)

# só mostrar agoraaa
plt.show()
```

## Seaborn
* Statistics plots - gráficos estatísticos
* Built on top of Matplotlib

```
# mostra uma noçãozinha de base
data_heart.head()
```

```
# importa a lib, ou seja, seaborn, que ´e quem a gente quer usar
import seaborn as sns
```

```
#displot - veja como é rodando a célula (POR PADRÃO, sem informar o kind, ele sai assim com esses retângulos bonitinhos). como ficam os dados de idade?
# o hist, histograma é o padrão, esses retângulos.
sns.displot(data_heart['age'])
```

```
# informando o kind, a gente pode variar a representação desses dados de muitos jeitos
sns.displot(data_heart['age'], kind='kde')
#sns.displot(data_heart['age'], kind='ecdf')
```

```
# quero a  minha representação padrão, mas quero um kde por cima? só adicionar kde=True
#sns.displot(data_heart['age'], kde=True)

# quero a mesma coisa, mas com o rug por cima?
sns.displot(data_heart['age'], rug=True)
```

```
# o bins passa quantos intervalos irão ter a representação. Falando de uma forma beem bruta, ele vai dar 15 retângulos (dividir os dados nisso), nem mais, nem menos.
# só dá pra fazer essa quantização com kind='hist', ou seja, os retângulos, o padrão que não precisa colocar nada.
sns.displot(data_heart['age'], bins=15)
```

```
# posso trocar a cor desses retângulos, escolher a altura do plot, que será de 8 unidades.
sns.displot(data_heart['age'], bins=20, color='r', height=8)
```

```
# mostra
data_heart.head()
```

```
# agora a gente vai usar o countplot, plotar de acordo com quantidade(?)
# exatamente, contagem de valores categóricos
# sns.countplot(data=nome_da_base, x='nome_da_coluna')
sns.countplot(data=data_heart, x='sex')
#perceba que ele está contando o número de zeros, o número de uns.
```

```
# agora com uma coluna com dados não binários
sns.countplot(data=data_heart, x='cp')
#criar uma label pra fazer uma graça
plt.xlabel('Chest Pain Type')
```

```
# o hue não adiciona somente uma legenda. você pode adicionar a coluna  para o "hue" que deseja
# o que está acontecendo aqui? o hue(sexo) está mostrando sua quantização para cada cp(0,1,2 e 3)
sns.countplot(data=data_heart, x='cp', hue='sex')
```

```
# mostra
data_heart.head()
```

```
# boxplot, saímos de countplot. Só passar a coluna como uma variável y, pra ela ficar em pézinha. Se passar como x, ela fica deitada, saca?
sns.boxplot(data=data_heart, y='age')
```

```
# boxplot de idade por sexo
sns.boxplot(data=data_heart, y='age', x='sex')
```

```
# mostra
data_heart.head()
```

```
sns.boxplot(data=data_heart, y='chol', x='target')
```

```
# tá pegando o chol, classificando por target e ainda classificando por sexo
sns.boxplot(data=data_heart, y='chol', x='target', hue='sex')
```

```
# chol x trestbps
sns.scatterplot(data=data_heart, x='chol', y='trestbps')
```

```
# os dados chol x trestbps são separados de acordo com sua target (os seus dados são binários, então fica assim, certo?)
sns.scatterplot(data=data_heart, x='chol', y='trestbps', hue='target')
```
