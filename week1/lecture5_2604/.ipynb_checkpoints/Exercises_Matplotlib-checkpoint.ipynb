Follow the instructions to recreate the plots using this data:

## Data

```python
# começando, óbvio, importando o numpy que é a grande base.
import numpy as np
# aqui eu vou criar uma array que vai de 0 a 20 (20 - 1)
x = np.arange(0,20)
x
# agora eu disse que x seria o quadrado dele mesmo agora + 1000, ou seja, cada elemento de x será elevado ao quadrado e somado a 1000
y = x**2 + 1000
y
# aqui eu irei inventar mais: quero que a minha nova array (z) seja o cubo de x - o quadrado de x 
z = x**3 - x**2
z
```

**1. Import matplotlib.pyplot as plt**

```python
# importabdi a lib que é o foco de agora
import matplotlib.pyplot as plt
```

**2. Plot (x,y), set the x-axis label to 'x', set the y-axis label to 'y' and set the title to 'TITLE' as in the figure below.**



image.png

```python
# PLOTAR 
plt.plot(x, y)
#BOTAR UM TÍTULO
plt.title('TITLE')
#colocar um rótulo na coordenada x
plt.xlabel('x')
# colocar um rótulo na coordenada y
plt.ylabel('y')
```

**3. Now plot (x, z) to recreate the plot below. Observe the labels and limits of the axis on the plot.**

image.png

```python
# plotando o gráfico
plt.plot(x, z)
#botando o título
plt.title('TITLE')
# rotulando a coordenada x
plt.xlabel('x')
# estabelecendo os limites de x
plt.xlim(left=0, right=16)
# rotulando a coordenada y
plt.ylabel('y')
# estabelecendo os limites de y
plt.ylim(bottom=0, top=4000)
```

**4. Plot (x, y) and (x, z) as in the plot below. Notice the legend.**

image.png

```python
# plotando os dois de uma vez
plt.plot(x, y, x, z)
#botando o título
plt.title('TITLE')
# botar a legenda de cada função
plt.legend(['x² + 1000', 'x³ - x²'])
```

**5. Use plt.subplots(nrows=1, ncols=2) to create the plot below. Note the colors and line styles.**

image.png

```python
#criando um subplot

plt.subplot(2, 1, 1)
plt.plot(x, y, 'r--')
plt.ylabel('y')
plt.subplot(2, 1, 2)
plt.plot(x, z, 'b')
plt.ylabel('z')

plt.xlabel('x')

```

**6. Resize the plot to (5, 15) as in the figure below.**

image.png

```python
plt.figure(figsize=(5,15))
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r--')
plt.ylabel('y')
plt.subplot(2, 1, 2)
plt.plot(x, z, 'b')
plt.ylabel('z')

plt.xlabel('x')
```
