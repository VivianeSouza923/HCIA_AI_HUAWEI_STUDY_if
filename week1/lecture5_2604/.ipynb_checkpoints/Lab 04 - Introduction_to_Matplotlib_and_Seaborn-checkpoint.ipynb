## Matplotlib
* Plotting library for Python 

```python
# importando as libs que eu vou precisar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# criando uma array com intervalo de 1 a 50
x = np.arange(1, 50)
x
```

```python
# criando uma nova array com o log da array x
y = np.log(x)
y
```

```python
# vamos ver isso em gráfico com a lib Matplotlib
plt.plot(x, y)

# isso aqui já é autoexplicativo!
plt.show()
```

```python
# cria agora uam array com o intervalo -20 a 20
x2 = np.arange(-20, 20)
x2
```

```python
# faz algumas operações com o x2 e atribuindo a y2
y2 = 0.5 * x2 ** 2 + 0.5* x2 + 0.5
y2
```

```python
plt.plot(x2, y2)
```

```python
# no plot passado, ficou azul, né? mas se eu quiser pink/mangenta?
# só acrescentar 'm'
plt.plot(x2, y2, 'm')
```

```python
# agora se eu quiser essa curva pontilhada? 
# é só 'm--' acrescenta dois tracinhos ao lado da letra
plt.plot(x, y, 'm--')
```

```python
# aqui eu posso definir o tamanho da figura
plt.figure(figsize=(10, 6))

# faz o plot
plt.plot(x, y, 'y', x2, y2, 'g--')
```

```python
# aqui eu tô plotando plt.plot(primeiras coordenadas, 'cor', segundas coordenadas, 'cor com tracejado')
plt.plot(x, y, 'r', x2, y2, 'b--')
# quero colocar a legenda na figura. Para isso eu passo uma lista e tals
plt.legend(['log', 'x**2'])
```

```python
# aqui eu tô danod o plot, pra aparecer o gráfico e tal com o x e y(lembrando que o x é uma array que criamos de 
# 1 a 50 e o y é o log da array x)
plt.plot(x, y)
# quero colocar um rótulo na coordena x : plt.xlabel('rótulo')
plt.xlabel('x')
# quero colocar um rótulo na coordena y : plt.ylabel('rótulo')
plt.ylabel('log(x)')
# agora eu quero um título também. plt.title('nome que eu quero dar', fonte que eu quero=x). lembrando que a fonte
# é "fontsize"
plt.title('Plot', fontsize=20)
# agora é só exibir
plt.show()
```

###Subplots

```python
# tamanho da figura
plt.figure(figsize=(15, 5))
# bora plotar as coisas agora. 
#plt.subplot(número de linhas, número de colunas  - cada figura será uma coluna, index)
plt.subplot(1, 2, 1)
# plota o primeiro gráfico
plt.plot(x, y, 'r')
# estabelece o limite para o x. NO caso, aqui vai 0 a 60
plt.xlim(right=60)
# estabelece o limite para y da figura 1
plt.ylim(bottom=-5, top=8)
plt.xticks(np.arange(0, 61, 5))


plt.subplot(1, 2, 2)
plt.plot(x2, y2, 'b--')
plt.xlim(left=-30, right=60)



plt.show()
```
