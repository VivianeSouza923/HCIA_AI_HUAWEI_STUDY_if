# Visualization Exercise

## The Data

This classic dataset contains the prices and other attributes of almost 54,000 diamonds. It's a great dataset for beginners learning to work with data analysis and visualization.

https://www.kaggle.com/shivam2503/diamonds

#### Content

* price price in US dollars (\$326--\$18,823)

* carat weight of the diamond (0.2--5.01)

* cut quality of the cut (Fair, Good, Very Good, Premium, Ideal)

* color diamond colour, from J (worst) to D (best)

* clarity a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

* x length in mm (0--10.74)

* y width in mm (0--58.9)

* z depth in mm (0--31.8)

* depth total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)

* table width of top of diamond relative to widest point (43--95)

**1. Import pandas, matplotlib.pyplot and seaborn**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**2. Read in the data set diamonds.csv**

```python
df_diamond = pd.read_csv('diamonds.csv')
df_diamond
```

**3. Create a scatterplot of `price` versus `carat` as shown below.**

image.png

```python
# sns.scatterplot(data=nome_da_base, x='o que eu quero de dados', y='o que eu quero de dados')
sns.scatterplot(data=df_diamond, x='carat', y='price')
```

**4. Make the previous plot larger with figure size (10, 8).**




image.png

```python
sns.scatterplot(data=df_diamond, x='carat', y='price', size_norm=(10,8))
```

**5. Create a histogram of the `price` column with `displot` as shown below. Observe the x-axis limits. Also, set bins=50 and height=8.**

image.png

```python
sns.displot(df_diamond['price'], bins=50, height=8)
plt.xlim(0,18000)
```

**6. Create a count plot of the instances per `cut` type as shown below.**

image.png

```python
sns.countplot(df_diamond, x='cut')
```

**7. Create a large box plot figure showing the `price` distribution per `cut` type as shown below. Set (10, 8) as figure size.**

image.png

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df_diamond, x='cut', y='price')
plt.show()
```

**8. Figure out how to change the ordering of the box plot as shown below.**

image.png

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df_diamond, x='cut', y='price', order=['Fair', 'Ideal', 'Good', 'Very Good', 'Premium'])
plt.show()
```
