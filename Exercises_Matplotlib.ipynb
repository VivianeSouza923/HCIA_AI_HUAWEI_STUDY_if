Follow the instructions to recreate the plots using this data:

## Data

```python
# importar o que eu vou precisar para criar array:
import numpy as np
x = np.arange(0,20)
y = x**2 + 1000
z = x**3 - x**2
```

**1. Import matplotlib.pyplot as plt**

```python
import matplotlib.pyplot as plt
```

**2. Plot (x,y), set the x-axis label to 'x', set the y-axis label to 'y' and set the title to 'TITLE' as in the figure below.**



image.png

```python
# plotar x e y
plt.plot(x,y)
#label to x
plt.xlabel("x")
#label to y
plt.ylabel("y")
# t√≠tulo
plt.title("TITLE")

#pra ver de uma forma mais bonita e elegante:
plt.show()


```

**3. Now plot (x, z) to recreate the plot below. Observe the labels and limits of the axis on the plot.**

image.png

```python
# PLOTAR X E Z
plt.plot(x,z)
#limitar x
plt.xlim(0,16)
#limitar y
plt.ylim(0,4000)

# botei o r√≥tulo de x
plt.xlabel("x")
# botei o r√≥tulo de y
plt.ylabel("z")
# botei t√≠tulo
plt.title("TITLE")

plt.show()
```

**4. Plot (x, y) and (x, z) as in the plot below. Notice the legend.**

image.png

```python
# plotar as duas fun√ß√µes numa figura s√≥
plt.plot(x, y, x, z, 'orange')
# r√≥tulo de x
plt.xlabel("x")

# t√≠tulo
plt.title("TITLE")

# colocar essas legendas naquele cantinho de cima, sabe?
plt.legend(["x¬≤+1000", "x¬≥-x¬≤"])

# mostrar de uma forma mais bonita e elegante
plt.show()
```

**5. Use plt.subplots(nrows=1, ncols=2) to create the plot below. Note the colors and line styles.**

image.png

```python
# primeiro tem que subplotar.
# plt.subplot(numero_De_linhas, numero_De_colunas, index)
plt.subplot(2,1,1)
#depois plota normal
plt.plot(x, y, 'r--')
#p√µe o r√≥tulo
plt.ylabel('y')
# subplot de novo, com o index 2 agora
plt.subplot(2,1,2)
# depois plota normal de novo
plt.plot(x,z,'b')
# p√µe os r√≥tulos
plt.xlabel('x')
plt.ylabel('z')


# pra ficar mais elegante
plt.show()












```
