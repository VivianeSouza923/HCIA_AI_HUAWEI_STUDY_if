```python
# importar o que eu vou precisar para criar array:
import numpy as np
x = np.arange(0,20)
y = x**2 + 1000
z = x**3 - x**2
```

**1. Import matplotlib.pyplot as plt**

```python
import matplotlib.pyplot as plt
```

**2. Plot (x,y), set the x-axis label to 'x', set the y-axis label to 'y' and set the title to 'TITLE' as in the figure below.**



image.png

```python
# plotar x e y
plt.plot(x,y)
#label to x
plt.xlabel("x")
#label to y
plt.ylabel("y")
# t√≠tulo
plt.title("TITLE")

#pra ver de uma forma mais bonita e elegante:
plt.show()


```

**3. Now plot (x, z) to recreate the plot below. Observe the labels and limits of the axis on the plot.**

image.png

```python
# PLOTAR X E Z
plt.plot(x,z)
#limitar x
plt.xlim(0,16)
#limitar y
plt.ylim(0,4000)

# botei o r√≥tulo de x
plt.xlabel("x")
# botei o r√≥tulo de y
plt.ylabel("z")
# botei t√≠tulo
plt.title("TITLE")

plt.show()
```

**4. Plot (x, y) and (x, z) as in the plot below. Notice the legend.**

```python
# plotar as duas fun√ß√µes numa figura s√≥
plt.plot(x, y, x, z, 'orange')
# r√≥tulo de x
plt.xlabel("x")

# t√≠tulo
plt.title("TITLE")

plt.legend(y)

plt.show()
```
